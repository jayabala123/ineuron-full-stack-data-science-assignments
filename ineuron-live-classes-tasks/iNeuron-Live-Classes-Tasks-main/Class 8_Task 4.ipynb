```python
# importing libraries
import cv2
import numpy as np

def read_play_video(cap):
  if (cap.isOpened()== False): # Check if camera opened successfully
   print("Error opening video file")
  
  while(cap.isOpened()): # Read until video is completed
   ret, frame = cap.read() # Capture frame-by-frame
   if ret == True:
     cv2.imshow('Frame', frame) # Display the resulting frame
     if cv2.waitKey(25) & 0xFF == ord('q'): # Press Q on keyboard to exit
      break
   else:
    break # Break the loop
  
  cap.release() # When everything done, release # the video capture object
  cv2.destroyAllWindows() # Closes all the frames

# Function Call by Creating a VideoCapture object and read from input file
read_play_video(cv2.VideoCapture('test.mp4'))
```
